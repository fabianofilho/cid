```
import pandas as pd
```

```
cid = pd.read_csv('cid10.csv',  index_col=0)
```

```
cid
```

```
dicid = cid.to_dict()
```

```
import json
json = json.dumps(dicid)
f = open("dicid.json","w")
f.write(json)
f.close()
```

```
import pickle
f = open("dicid.pkl","wb")
pickle.dump(dicid,f)
f.close()
```
