```
mortalidade = pd.read_csv('http://bit.ly/df_sim_es_2017')
```

```
cid_mort = mortalidade[['NUMERODO', 'LINHAA', 'LINHAB', 'LINHAC', 'LINHAII', 'CAUSABAS']]
```

```
cid_mort.isnull().sum()
```

```
cid_mort.dropna(inplace=True)
```

```
cid_mort["linhaA"] = cid_mort["LINHAA"].str.split("*") 
cid_mort["linhaB"] = cid_mort["LINHAB"].str.split("*") 
cid_mort["linhaC"] = cid_mort["LINHAC"].str.split("*") 
cid_mort["linhaII"] = cid_mort["LINHAII"].str.split("*") 
cid_mort["causabas"] = cid_mort["CAUSABAS"].str.split("*") 
```

```
#cid_mort
```

```
cid_mort.head(2)
```

```
cid_morte = cid_mort[['NUMERODO',	'linhaA',	'linhaB',	'linhaC',	'linhaII',	'causabas']]
```

```
cid_morte = cid_morte[ cid_mort['LINHAA'] != cid_mort['CAUSABAS'] ]
```

```
cid_morte.shape
```

```
cid_caus_consq = cid_morte[['causabas', 'linhaA']]
```

```
#cid_caus_consq['linhaA'].tolist()
```

```
def unpack(df, column, fillna=None):
    ret = None
    if fillna is None:
        ret = pd.concat([df, pd.DataFrame((d for idx, d in df[column].iteritems()))], axis=1)
        del ret[column]
    else:
        ret = pd.concat([df, pd.DataFrame((d for idx, d in df[column].iteritems())).fillna(fillna)], axis=1)
        del ret[column]
    return ret

unpack(cid_caus_consq, 'linhaA', 0)
```

```
items_as_cols = cid_caus_consq.apply(lambda x: pd.Series(x['causabas']), axis=1)

# Keep original df index as a column so it's retained after melt
items_as_cols['orig_index'] = items_as_cols.index

melted_items = pd.melt(items_as_cols, id_vars='orig_index', 
                       var_name='caus_num', value_name='caus')
melted_items.set_index('orig_index', inplace=True)

cid_caus_consq1 = cid_caus_consq.merge(melted_items, left_index=True, right_index=True)
```

```
items_as_cols = cid_caus_consq1.apply(lambda x: pd.Series(x['linhaA']), axis=1)

# Keep original df index as a column so it's retained after melt
items_as_cols['orig_index'] = items_as_cols.index

melted_items = pd.melt(items_as_cols, id_vars='orig_index', 
                       var_name='A_num', value_name='A')
melted_items.set_index('orig_index', inplace=True)

cid_caus_consq2 = cid_caus_consq1.merge(melted_items, left_index=True, right_index=True)
```
